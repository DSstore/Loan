```python
from autogluon.tabular import TabularDataset, TabularPredictor
import pprint
from IPython.display import display
import pandas as pd
```

```python
main_data = TabularDataset("https://github.com/DSstore/Loan/raw/main/feature_engineering_train.csv")
main_data
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(main_data, train_size=0.8, test_size=0.2, random_state=1)
```

```python
train_data
```

```python
test_data
```

```python
train_data.describe()
```

```python
label_column = 'Loan_Status'
print('Number of Possible result of target variable\n', main_data[label_column].value_counts().to_string())
```

```python
metric = 'roc_auc'  # specify your evaluation metric here
predictor = TabularPredictor(label_column, eval_metric=metric).fit(train_data, presets='best_quality')
predictor.leaderboard(test_data, silent=True)
```

```python
y_test = test_data[label_column]  # values to predict
test_data_nolab = test_data.drop(columns=[label_column])  # delete label column to prove we're not cheating
test_data_nolab.head()
```

```python
print("AutoGluon infers problem type is: ", predictor.problem_type)
print("AutoGluon identified the following types of features:")
print(predictor.feature_metadata)
```

```python
pred_probs = predictor.predict_proba(test_data, model='RandomForestGini_BAG_L1')
pred_probs.head(5)
```

```python
y_pred = predictor.predict(test_data_nolab, model='RandomForestGini_BAG_L1' )
print("Predictions:  \n", y_pred)
perf = predictor.evaluate_predictions(y_true=y_test, y_pred=pred_probs, auxiliary_metrics=True )
```

```python
predictor.leaderboard(test_data, silent=True)
```

```python
predictor.feature_importance(test_data, model='RandomForestGini_BAG_L1')
```

```python
testing_data = TabularDataset("https://github.com/DSstore/Loan/raw/main/feature_engineering_test.csv")
testing_data_pred = predictor.predict(testing_data, model='RandomForestGini_BAG_L1')
testing_data_pred
```
